```python
import numpy as np
import matplotlib.pyplot as plt

# 실제 데이터 생성
np.random.seed(0)  # 난수 발생을 위한 시드 설정
data = np.random.normal(loc=0, scale=1, size=1000)  # 평균이 0이고 표준편차가 1인 정규 분포에서 1000개의 샘플 생성

# 데이터의 표준편차 계산
std_dev = np.std(data)

# 히스토그램을 이용한 데이터 분포 시각화
plt.figure(figsize=(8, 6))  # 그래프 크기 지정
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')  # 히스토그램 생성
plt.title('Histogram of Data')  # 그래프 제목
plt.xlabel('Value')  # x 축 레이블
plt.ylabel('Frequency')  # y 축 레이블
plt.grid(True)  # 그리드 추가

# 표준편차를 수직선으로 표시
plt.axvline(x=std_dev, color='red', linestyle='--', linewidth=2, label=f'Standard Deviation: {std_dev:.2f}')
plt.axvline(x=-std_dev, color='red', linestyle='--', linewidth=2)

# 범례 추가
plt.legend()

# 그래프 표시
plt.show()
```
